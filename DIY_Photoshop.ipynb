Import python modules. These are libraries that will allow us to read in images and perform math operations.

```python
import numpy as np
from PIL.Image import open
import matplotlib.pyplot as plt
import cv2 as cv
```

Open an image

```python
I = np.array(open('Iribe.jpg'))/255
```

Plot the image

```python
plt.imshow(I)
plt.show()
```

Check how large the image is

```python
print(I.shape)
```

Visualize the red channel

```python
plt.imshow(I*np.array([1,0,0]))
plt.show()
```

Visualize the green channel

```python
plt.imshow(I*np.array([0,1,0]))
plt.show()
```

Visualize the blue channel

```python
plt.imshow(I*np.array([0,0,1]))
plt.show()
```

Load and display a second image
